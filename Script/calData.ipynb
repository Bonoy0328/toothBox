```python
import os
from matplotlib import pyplot as plt
f = open("data.txt",'r')
data = f.read()
buf = []
for i in data.split("\n"):
    buf.append(float(i.split("=")[1]))
```

```python
plt.figure(figsize=(40,5))
plt.grid()
plt.plot(buf)
plt.plot(buf,'ro')
```

```python
# 100
fil = [65535] * 100
buffil = []
for i in buf:
    fil.append(i)
    del fil[0]
    buffil.append(sum(fil) / 100 )
plt.figure(figsize=(40,5))
plt.grid()
plt.plot(buf[100:])
plt.plot(buf[100:],'ro')
plt.plot(buffil[100:],'g')
```

```python
shled = 0.08
buffiledvalue = []
buffiledindex = []
for i in range(len(buf)):
    if(buf[i] > (buffil[i] * (1 + shled))):
        if(i > 100):
            buffiledvalue.append(buf[i])
            buffiledindex.append(i-100)
plt.figure(figsize=(40,5))
plt.grid()
plt.plot(buf[100:])
plt.plot(buf[100:],'ro')
plt.plot(buffil[100:],'g')
plt.plot(buffil[100:],'g')
plt.plot(buffiledindex,buffiledvalue)
```
